```python

def filter_excel_file(file_path, filter_size=None, filter_name=None):
   

    df = pd.read_excel("shoe_store.xlsx")
    # Create a copy
    original_df = df.copy()

    # Apply filters
    if filter_size:
        df = df[df['Size'] == filter_size]
    if filter_name:
        df = df[df['Name'] == filter_name]

    # Reset the index of the filtered DataFrame
    df = df.reset_index(drop=True)

    # Assign the original index values as the 'Index' column
    df['Index'] = original_df.index[df.index]

    # Apply styling to the filtered DataFrame
    styled_df = (
        df.style
        .set_properties(subset=['Index'], **{'text-align': 'center', 'font-weight': 'bold'})
        .background_gradient(subset=['Price'], cmap='YlGn', low=0, high=500)
        .highlight_max(subset=['Size'], color='lightgreen')
        .highlight_min(subset=['Size'], color='lightcoral')
        .format({'Price': '${:,.2f}', 'Size': '{:.0f}'})
    )

    # Convert styled DataFrame to a regular DataFrame
    regular_df = styled_df.data

    # Generate a visual representation of the DataFrame using matplotlib
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.axis('off')
    ax.table(cellText=regular_df.values, colLabels=regular_df.columns, loc='center', cellLoc='center')
    plt.show()

    # Return the styled DataFrame and the original DataFrame
    return styled_df, original_df


styled_df, original_df = filter_excel_file('shoe_store.xlsx', filter_size=40)
```

```python
import pandas as pd

# Function to filter Excel by stock number
def filter_excel_by_stock(excel_path, stock_number):
    df = pd.read_excel(excel_path)
    filtered_df = df[df['Stock Number'] == stock_number]
    return filtered_df

# Function to delete a specific stock number from the Excel
def delete_stock_number(excel_path, stock_number):
    df = pd.read_excel(excel_path)
    df = df[df['Stock Number'] != stock_number]
    df.to_excel(excel_path, index=False)

# Function to update stock number by location
def update_stock_by_location(excel_path, location, stock_number):
    df = pd.read_excel(excel_path)
    df.loc[df['Location'] == location, 'Stock Number'] = stock_number
    df.to_excel(excel_path, index=False)

# Example usage
kids_excel_path = 'kids.xlsx'
women_excel_path = 'women.xlsx'
men_excel_path = 'men.xlsx'


```
